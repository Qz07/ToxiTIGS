```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel

device = "cuda" if torch.cuda.is_available() else "cpu"

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2").to(device)
model.eval()
```

```python
import random
```

```python
idx = random.choice(range(len(ds['train'])))
ds['train'][idx]
```

```python
pmt = ds['train'][idx]['prompt']
```

```python
len(tokenizer.encode(pmt)), len(tokenizer.encode(ds['train'][idx]['generation']))
```

```python
inputs = tokenizer(
    pmt,
    return_tensors="pt"
).to(device)

with torch.no_grad():
    generated_ids = model.generate(
        **inputs,
        max_length=256,
        do_sample=False
    )

print(tokenizer.decode(generated_ids[0], skip_special_tokens=True).replace(pmt, ''))
```

```python
import pickle 
from tqdm import tqdm
```

```python
with open('../data/jan13_toxic_data_250k.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
data[0]
```

```python
pos = [] # label 0
neg = [] # label 1
for ele in tqdm(data):
    if ele['label'] == 1:
        neg.append(ele)
    else:
        pos.append(ele)
```

```python
len(pos), len(neg)
```

```python
all_prompt = {}
for ele in tqdm(data):
    if ele['prompt'] not in all_prompt:
        all_prompt[ele['prompt']] = []
    all_prompt[ele['prompt']].append(ele)
```
