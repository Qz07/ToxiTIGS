```python
from datasets import load_dataset
from tqdm import tqdm
```

```python
ds = load_dataset("google/civil_comments", split='test', streaming=False)
```

```python
ds
```

```python
ds[0]
```

```python
toxic_lst = []
for ele in tqdm(ds):
    if ele['toxicity'] > 0.5:
        toxic_lst.append(ele)
```

```python
len(toxic_lst)
```

```python
import random
```

```python
random.seed(42)
samp1000 = random.sample(toxic_lst, 1000)
```
