```python
import pickle
from tqdm import tqdm
```

```python
with open('../data/jan13_toxic_data_250k.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_path = "gpt2"                 # or "/path/to/your/checkpoint_dir"
tok = AutoTokenizer.from_pretrained(model_path, use_fast=True)

# GPT-2 often has no pad token by default; set one if you batch/pad
if tok.pad_token is None:
    tok.pad_token = tok.eos_token
```

```python
seq_len = []
lt_256 = []
for ele in tqdm(data):
    total_str = ele['prompt']+ele['generation']
    cur_len = len(tok(total_str)['input_ids'])
    seq_len.append(cur_len)
    if cur_len < 256:
        lt_256.append(ele)
```

```python
max(seq_len), min(seq_len), sum(seq_len)/len(seq_len)
```
